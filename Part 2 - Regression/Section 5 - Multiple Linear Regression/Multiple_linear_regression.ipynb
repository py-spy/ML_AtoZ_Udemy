This data contains information on 50 Startups with Amount of money spent in Admin R&D etc. We have a profit column. We will build a model to see if there is any linear dependncy between variables and profit. Profit is the target. There are investors who want to know in which startup they should invest their money. They also want to see which independent variable which has maximum effect.

```python
# Data Preprocessing Template
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
# Importing the dataset
dataset = pd.read_csv('50_Startups.csv')

dataset
```

```python
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:,4].values
```

```python
# Encoding categorical data
# Encoding the Independent Variable
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
labelencoder_X = LabelEncoder()
X[:, 3] = labelencoder_X.fit_transform(X[:, 3])
onehotencoder = OneHotEncoder(categorical_features = [3])
X = onehotencoder.fit_transform(X).toarray()
```

```python
X
```

```python
#Split
from sklearn.cross_validation import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.2, random_state = 0)
```

```python
#Fitting Mult linear regressor to training set
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train,y_train)
```

```python
#Predicting test results
y_pred = regressor.predict(X_test)
```

```python
# Building the optimal model using Backward Elimination
# removing non statistically significant variables
import statsmodels.formula.api as sm
# add columns of 1s to our matrix of indep variables (features)
X = np.append(arr = np.ones((50,1)).astype(int) , values = X, axis = 1 )
```

```python
X # 1s have been added to first column x(zero) = 1 of the mult lin reg equation
```

```python
X_opt = X[:,[0,1,2,3,4,5]] 
# if p value of indep variale is below significance 
#level it will stay, if p value of indep variable is above significance level
# then it will not stay in the model. SL = 0.05
regressor_OLS = sm.OLS(endog = y, exog = X_opt).fit() # Ordinary Least squares model
```

```python
regressor_OLS.summary() #lower p value more sig is the indep variable
```

```python
X_opt = X[:,[0,1,2,3,4]] 
regressor_OLS = sm.OLS(endog = y, exog = X_opt).fit()
regressor_OLS.summary()
```
